```python
from duckduckgo_search import *
from fastcore.all import *
import time

def search_images(name, max_images=200):
    url = 'https://duckduckgo.com/'
    res = urlread(url, data={'q':name})
    searchObj = re.search(r'vqd=([\d-]+)\&', res)
    requestUrl = url + 'i.js'
    params = dict(l='us-en', o='json', q=name, vqd=searchObj.group(1), f=',,,', p='1', v7exp='a')
    urls,data = set(),{'next':1}
    while len(urls)<max_images and 'next' in data:
        data = urljson(requestUrl, data=params)
        urls.update(L(data['results']).itemgot('image'))
        requestUrl = url + data['next']
        time.sleep(0.2)
    return L(urls)[:max_images]
```

```python
urls = search_images('bird photos', max_images=1)
urls[0]
```

```python
from fastdownload import download_url

dest = 'bird.jpg'
download_url(urls[0], dest, show_progress=True)

from fastai.vision.all import *

im = Image.open(dest)
im.to_thumb(256, 256)
```

- Now to do it with different setting of birds and forest normal, with sun and without sun

```python
searches = 'bird', 'forest'
path = Path('bird_or_not')

from time import sleep

for o in searches:
    dest = (path/o)
    dest.mkdir(exist_ok=True, parents=True)
    download_images(dest, urls=search_images(f"{o} photos"))
    sleep(10)
    download_images(dest, urls=search_images(f"{o} sun photos"))
    sleep(10)
    download_images(dest, urls=search_images(f"{o} shade photos"))
    sleep(10)
    resize_images(path/o, max_size=400, dest=path/o)
```
