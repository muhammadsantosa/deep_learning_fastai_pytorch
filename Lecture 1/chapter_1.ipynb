```python
from fastai.vision.all import *
path = untar_data(URLs.PETS)/'images'
```

```python
def is_cat(x): return x[0].isupper()
```

```python
dls = ImageDataLoaders.from_name_func(
    path, get_image_files(path), valid_pct=0.2, seed=42,
    label_func=is_cat, item_tfms=Resize(224))
```

- dls is used to tell fastai what type of dataset we have
- ImageDataLoaders loads image data
- get_image_files needs the path of all the images
- valid_pct is the validation size, so 20% of the training set.
    - Best practice to reduce overfitting
- label_func uses the is_cat as the labeler. This is based off of the parent folder that it is under
- item_tfms transforms the items(The images) into 224 x 224 sized pixels. This is a standard size but not required.
    - The higher the resolution, the more accurate the results, but the longer the training time

```python
learn = vision_learner(dls, resnet34, metrics=error_rate)
```

- Creates a vision learner, that passes in the dls data, utilises a pretrained model of resnet34, and uses error rate as the performance metric
- resnet34 is a pretrained model made for computer vision, with convolutional neural networks specialising
    - 34 represents the number of layers
- error rate as the metric is used for us humans to understand more closely
- a measure of performance is something that the stochastic gradient descent can use to optimize (loss function)
- vision_learner is automatically set to pretrained=True
- Pretrained models are very important, and is BEST PRACTICE
- Transfer learning can help to use pretrained models for one task for another task


```python
learn.fine_tune(1)
```

```python
learn.show_results(max_n=6, figsize=(7,8))
```

- Fitting the model and fine tunning it for 1 epoch
- This changes the head of the model which when using pretrained weights, is the most important part to change

What Our Image Recognizer Learned
- Different layers have different things that they are looking for in a CNN
- Resembles closely to how they human eye works

- There was a case of transfer learning, using a cnn or resnet architecture to categorize sound spectograms into their respective sounds

Summary

- 1 training 'generation' is called an epoch

Training a Semantic Segmentation Model with Fastai

```python
path = untar_data(URLs.CAMVID_TINY)
```

```python
dls = SegmentationDataLoaders.from_label_func(
    path, bs=8, fnames = get_image_files(path/'images'),
    label_func = lambda o: path/'labels'/f'{o.stem}_P{o.suffix}',
    codes = np.loadtxt(path/'codes.txt', dtype=str)
)
```

- bs is batch size, so it does segmentation in batches
- I assume f_names is just the name of the files
- label_func is the same as the is_cat function
- codes is the mapping index to the label

```python
learn = unet_learner(dls, resnet34) 
```

```python
learn.fine_tune(8)
```

```python
learn.show_results(max_n=2, figsize=(7,8))
```

Natural Language Processing for Sentiment Analysis of Movies

```python
from fastai.text.all import *

dls = TextDataLoaders.from_folder(untar_data(URLs.IMDB), valid='test')
```

```python
learn = text_classifier_learner(dls, AWD_LSTM, drop_mult=0.5, metrics=accuracy)
```

- Text classifier that takes in a dataloader, uses a pretrained LTSM model, drop_mult need to check, and uses the accuracy metric

```python
learn.fine_tune(4, 1e-2) #Just as a warning, dont run this, takes too long
```

Training a Model for Tabular Data

```python
from fastai.tabular.all import *
path = untar_data(URLs.ADULT_SAMPLE)
```

```python
dls = TabularDataLoaders.from_csv(path/'adult.csv', path=path, y_names='salary',
                                 cat_names = ['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race']
                                 cont_names = ['age', 'fnlwgt', 'education-num'],
                                 procs = [Categorify, FillMissing, Normalize])
```

```python
learn = tabular_learner(dls, metrics=accuracy)
```

```python
learn.fit_one_cycle(3) #Fit one cycle is usually used when no pretrained model is being used
```
